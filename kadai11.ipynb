```python
#for構文を用い、リストの中身を先頭から全て取り出し、それを反復する

user_names = ["水","金","地","火","木","土","天","海","冥"]

for user_name in user_names:
  print(user_name)

for user_name in user_names:
  print(user_name)

```

```python
#while構文を用い、リストの中身を先頭から全て取り出し、それを反復する

user_names = ["水", "金", "地", "火", "木", "土", "天", "海", "冥"]
count = 0

while count < 2:
    for user_name in user_names:
      print(user_name)
    count += 1
```
