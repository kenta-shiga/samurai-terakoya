```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name}は大人です。")
        else:
            print(f"{self.name}は大人ではありません。")

# Humanクラスのインスタンスを複数生成してリストに追加
humans = [
    Human("侍太郎", 25),
    Human("侍花子", 18),
    Human("侍健太郎", 50)
]

# 各インスタンスのcheck_adultメソッドを呼び出す
for human in humans:
    human.check_adult()
```
