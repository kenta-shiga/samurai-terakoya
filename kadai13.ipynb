```python
def add_two_arguments(product_price, tax_rate):
    if product_price < 0 or tax_rate < 0 or tax_rate > 1:
        return "正しい金額と税率を入力してください。"

    total_price = product_price * (1 + tax_rate)
    return total_price

# 例として商品の金額が1000円の場合の合計金額を計算
product_price = 1000
tax_rate = 0.10  # 10% の消費税
total_price = add_two_arguments(product_price, tax_rate)
print(f"商品の金額: {product_price} 円")
print(f"消費税率: {tax_rate * 100}%")
print(f"合計金額: {total_price} 円")

```
