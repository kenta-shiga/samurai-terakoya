```python
import openpyxl
from openpyxl.utils import get_column_letter

# 新規ワークブックを作成
wb = openpyxl.Workbook()
sheet = wb.active

# datetimeモジュールをインポート
import datetime
now = datetime.datetime.now()

# タイトルと企業情報の設定
sheet['B2'] = "請求書"
sheet['B4'] = "株式会社ABC"
sheet['B5'] = "〒101-0022 東京都千代田区神田練塀町300"
sheet['B6'] = "TEL:03-1234-5678 FAX:03-1234-5678"
sheet['B7'] = "担当者名:鈴木一郎 様"

sheet['F4'] = "No."
sheet['F5'] = "日付"
sheet['G4'] = "0001"

# 現在の日付と時刻をセルに設定
sheet['G5'] = now.strftime("%Y/%m/%d")

# ヘッダーの設定
header_labels = [ "商品名", "数量", "単価", "金額"]
for col_num, label in enumerate(header_labels, start=2):
    sheet[get_column_letter(col_num) + '10'] = label

# 商品データの設定
product_data = [
["商品A",2,10000],
["商品B",1,15000]
]

total = 0

for i , data in enumerate(product_data):
 price = data[1]*data[2]
 sheet.cell(row = i+11, column = 2).value = data[0]
 sheet.cell(row = i+11,column = 3).value = data[1]
 sheet.cell(row = i+11,column = 4).value = data[2]
 sheet.cell(row = i+11,column = 5).value = price
 total = total + price

last_row = sheet.max_row
sheet.cell(row = last_row + 1,column=5).value =  total

last_row = sheet.max_row
sheet.cell(row = last_row + 2,column=2).value = "小計"
sheet.cell(row = last_row + 2,column=5).value = total

sheet.cell(row = last_row + 3,column=2).value = "消費税"
sheet.cell(row = last_row + 3,column=5).value = total * 0.1

sheet.cell(row = last_row + 4,column=2).value = "合計"
sheet.cell(row = last_row + 4,column=5).value = total * 1.1

# ファイル保存
file_name = f"請求書.xlsx"
wb.save(file_name)

```
