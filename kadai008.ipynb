```python
#変数varが3の倍数の場合はFizz、5の倍数の場合はBuzz、3と5の倍数の場合はFizzBuzz、それ以外の場合は変数varを出力する条件式を記述する
#ただし、変数varは正の整数

import random

num = random.randint(0,100)

print(num)

if num % 3 == 0 : print("Fizz")

elif num % 5 == 0 : print("Buzz")

elif num % 3 == 0 and num % 5 == 0:print("FizzBuzz")

else : print("var")
```
